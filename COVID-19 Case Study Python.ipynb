# Task 3 – Handling Missing Data

Identify missing values and replace them using forward filling suitable for time-series data.

```python
confirmed = confirmed.ffill()
deaths = deaths.ffill()
recovered = recovered.ffill()
```

Forward filling propagates the previous valid observation forward to fill missing values.

# Task 4 – Data Cleaning and Preparation

Replace blank values in Province/State column with “All Provinces”.

```python
confirmed["Province/State"] = confirmed["Province/State"].fillna("All Provinces")
deaths["Province/State"] = deaths["Province/State"].fillna("All Provinces")
recovered["Province/State"] = recovered["Province/State"].fillna("All Provinces")
```


# Task 5 Independent Dataset Analysis
### Task 5.1 – Analyze the peak number of daily new cases in Germany, France, and Italy.

```python
countries = ["Germany","France","Italy"]

country_data = confirmed[confirmed["Country/Region"].isin(countries)]

country_sum = country_data.groupby("Country/Region").sum().drop(columns=["Lat","Long"], errors="ignore")

# keep only numeric columns (dates)
country_sum = country_sum.select_dtypes(include="number")

daily_cases = country_sum.diff(axis=1)

peak_cases = daily_cases.max(axis=1)

peak_cases
```

# Task 6 Data Transformation
## Task 6.1 – Transform deaths dataset from wide format to long format

```python
deaths_long = deaths.melt(
    id_vars=["Province/State","Country/Region","Lat","Long"],
    var_name="Date",
    value_name="Deaths"
)

deaths_long["Date"] = pd.to_datetime(deaths_long["Date"], format="%m/%d/%y")
```

The melt() function converts the dataset from wide format to long format.
Each date column becomes a row with two new columns:
Date and Deaths. This makes the data easier to analyze and visualize over time.

## Task 6.2 – Total deaths reported per country

```python
total_deaths_country = deaths.groupby("Country/Region").sum().iloc[:,-1]

total_deaths_country.sort_values(ascending=False).head(10)
```

## Task 6.3 – Top 5 countries with highest average daily deaths

```python
# Group by country
deaths_country = deaths.groupby("Country/Region").sum()

# Keep only numeric columns (date columns)
deaths_country = deaths_country.select_dtypes(include="number")

# Calculate daily deaths
daily_deaths = deaths_country.diff(axis=1)

# Average daily deaths
avg_daily_deaths = daily_deaths.mean(axis=1)

# Top 5 countries
avg_daily_deaths.sort_values(ascending=False).head(5)
```

The data was grouped by country and the total deaths were calculated.
The difference between consecutive days was computed to estimate daily deaths.
Then the average daily deaths for each country were calculated and the top 5 countries with the highest averages were identified.

## Task 6.4 – Evolution of deaths in United States

```python
usa_deaths = deaths[deaths["Country/Region"] == "US"]

# remove non-date columns first
usa_deaths = usa_deaths.drop(columns=["Province/State","Country/Region","Lat","Long"], errors="ignore")

# sum all provinces
usa_total = usa_deaths.sum()

# convert index to datetime
usa_total.index = pd.to_datetime(usa_total.index)

# plot
plt.figure(figsize=(10,5))
plt.plot(usa_total.index, usa_total.values)

plt.title("Total Deaths Over Time in USA")
plt.xlabel("Date")
plt.ylabel("Deaths")
plt.xticks(rotation=45)

plt.show()
```

# Task 7 Data Merging
## Task 7.1 – Merge confirmed, deaths, and recovered datasets

```python
confirmed_long["Date"] = pd.to_datetime(confirmed_long["Date"], format="%m/%d/%y")
deaths_long["Date"] = pd.to_datetime(deaths_long["Date"], format="%m/%d/%y")
recovered_long["Date"] = pd.to_datetime(recovered_long["Date"], format="%m/%d/%y")

merged = confirmed_long.merge(
    deaths_long,
    on=["Province/State","Country/Region","Lat","Long","Date"]
)

merged = merged.merge(
    recovered_long,
    on=["Province/State","Country/Region","Lat","Long","Date"]
)

merged.head()
```

## Task 7.2 – Monthly sum of confirmed cases, deaths and recoveries

```python
merged["Date"] = pd.to_datetime(merged["Date"])

merged["Month"] = merged["Date"].dt.to_period("M")

monthly_data = merged.groupby(["Country/Region","Month"])[["Confirmed","Deaths","Recovered"]].sum()

monthly_data.head()
```
